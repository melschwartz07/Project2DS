```python
import string
from collections import OrderedDict
from collections import Counter
#Step 1: open and read the contents of the file.
myNovel = open('TheWoodKing.txt')
contents = myNovel.read()
print(contents)

#Step 2: string manipulation
#remove punctuation, all lower case
dictionary = OrderedDict(sorted(dict()))

with open('TheWoodKing.txt', 'r') as myNovel:
   for contents in myNovel:
       contents = contents.strip()
       contents = contents.lower()
       contents = contents.translate(contents.maketrans("", "", string.punctuation))
       words = contents.split(" ")
       print(contents)

#function to count word occurances
       for word in words:
           if word in dictionary:
               dictionary[word] = dictionary[word] + 1
           else:
               dictionary[word] = 1

#print each word and how many occurances
for key in list(dictionary.keys()):
    print(key, ":", dictionary[key])

#print the total amount of words in the novel
print("Total amount of words: ", len(dictionary))



```
